```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv("/Users/bualoydgreat/PERSONAL-TEMPS/optimized-traffic-light-computer-vision/runs/detect/train3/results.csv")

df.columns = df.columns.str.strip()

columns_to_plot = ['train/box_loss', 'train/cls_loss', 'train/dfl_loss',
                   'metrics/precision(B)', 'metrics/recall(B)',
                   'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
                   'val/box_loss', 'val/cls_loss', 'val/dfl_loss']

# Limit epochs to 100
df = df[df['epoch'] <= 100]

fig, axes = plt.subplots(2, len(columns_to_plot) // 2, figsize=(20, 10))


for i, column in enumerate(columns_to_plot):
    row = i // (len(columns_to_plot) // 2)
    col = i % (len(columns_to_plot) // 2)
    axes[row, col].plot(df['epoch'], df[column], linewidth=5)
    axes[row, col].set_title(column, fontsize=17)
    axes[row, col].set_xlabel("Epoch", fontsize=14)
    axes[row, col].set_ylabel("Value", fontsize=14)
    axes[row, col].tick_params(axis='both', which='major', labelsize=12)
    axes[row, col].grid(False)


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv("/Users/bualoydgreat/PERSONAL-TEMPS/optimized-traffic-light-computer-vision/runs/classify/train2/results.csv")

df.columns = df.columns.str.strip()

columns_to_plot = ['train/loss', 
                   'metrics/accuracy_top1', 
                   'metrics/accuracy_top5',
                   'val/loss']

# Limit epochs to 100
df = df[df['epoch'] <= 100]

num_plots = min(len(columns_to_plot), 10)
num_rows = 2
num_cols = 5

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 10))


if num_plots < num_rows * num_cols:
    axes = axes.flatten()


for i in range(num_plots):
    column = columns_to_plot[i]
    ax = axes[i]
    ax.plot(df['epoch'], df[column], linewidth=4)
    ax.set_title(column, fontsize=17)
    ax.set_xlabel("Epoch", fontsize=14) 
    ax.set_ylabel("Value", fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=12)
    ax.grid(False)


plt.tight_layout()
plt.show()
```
